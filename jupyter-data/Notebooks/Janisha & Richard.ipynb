```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/gdrive')
import os
os.chdir("gdrive/MyDrive/EskulML2026")

# Import library untuk mengolah data
import pandas as pd

# Import library untuk visualisasi data
import seaborn as sns
import matplotlib.pyplot as plt

# Import model Regresi Linear dari Scikit-Learn
from sklearn.linear_model import LinearRegression
```

STEP 2 – Memuat Dataset Iris

```python
# Mengambil data otomatis dari internet
df = pd.read_csv('Iris.csv')

df.head()
```

STEP 3 – Menentukan Variabel Input dan Target

```python
# Variabel input (X): Panjang mahkota bunga
X = df[['PetalLengthCm']]

# Variabel target (y): Lebar mahkota bunga
y = df['PetalWidthCm']
```

STEP 4 – Visualisasi Hubungan Data

```python
# Membuat grafik sebar (scatter plot)
plt.figure(figsize=(8, 5))

# Titik-titik data asli
plt.scatter(X, y)

# Memberi label pada grafik
plt.xlabel('Panjang Mahkota (cm)')
plt.ylabel('Lebar Mahkota (cm)')
plt.title('Hubungan Panjang dan Lebar Mahkota Bunga Iris')

# Menampilkan grid agar grafik mudah dibaca
plt.grid(True)

# Menampilkan grafik
plt.show()
```

STEP 5 – Membuat Model Regresi Linear

```python
# Membuat objek model regresi linear
model = LinearRegression()
```

STEP 6 – Melatih Model (Training)

```python
model.fit(X, y)
```

STEP 7 – Visualisasi Hasil Regresi

```python
# Membuat grafik hasil pembelajaran model
plt.figure(figsize=(8, 5))

# Menampilkan data asli
plt.scatter(X, y, label='Data Asli')

# Menampilkan garis regresi (hasil prediksi model)
plt.plot(X, model.predict(X), color='red', label='Garis Regresi')

# Memberi label dan judul
plt.xlabel('Panjang Mahkota (cm)')
plt.ylabel('Lebar Mahkota (cm)')
plt.title('Hasil Regresi Linear')

# Menampilkan legenda
plt.legend()
plt.grid(True)
plt.show()
```

STEP 8 – Melakukan Prediksi Data Baru

```python
# Data baru yang ingin diprediksi

panjang_baru = pd.DataFrame({'PetalLengthCm': [5.5]}) # panjang mahkota bunga dalam cm
# Melakukan prediksi menggunakan model
prediksi = model.predict(panjang_baru)

# Menampilkan hasil prediksi
print(f"Prediksi lebar mahkota untuk panjang 5.5 cm adalah {prediksi[0]:.2f} cm")
```
